Using mtcars.sqlite we'll execute and answer the following questions:

**1.1** Select unique values in the year column and order them in acsending order. How many unique values are there?

**1.2** Create a column with vehicle age values

**1.3** If a vehicle is older than 45 years, then label it as "45+". Otherwise, label it as "<45". Create a new column that takes these two values

**1.4** How many vehicles are in our dataset that are under 45 and 45 and over?

**1.5** List all unique vehicle name values that have 8 cylinders with horsepower that is greater than 125 but doesn't exceed 200


```python
conn = sqlite3.connect('../data/mtcars.sqlite')
pd.read_sql_query("SELECT name FROM  sqlite_schema WHERE  type ='table' AND name NOT LIKE 'sqlite_%'", conn)
```

```python
# 1 Select unique values in the year column and order them in acsending order.

df = pd.read_sql_query("SELECT * FROM results", conn)
df
```

```python
pd.read_sql_query("""SELECT DISTINCT year 
                  FROM results
                  ORDER BY year ASC""", conn)
```

```python
#  How many unique values are there?
pd.read_sql_query("""SELECT COUNT(DISTINCT year) 
                  FROM results
                  ORDER BY year ASC""", conn)
```

```python
# 1.2 Create a column with vehicle age values
df1 = pd.read_sql_query("""SELECT *, (2022-(1900+year)) AS age_veh FROM results """,conn)
df1
```

```python
# 1.3 If a vehicle is older than 45 years, then label it as "45+". 
# Otherwise, label it as "<45". Create a new column that takes these two values
df1 = pd.read_sql_query("""
                    SELECT *,
                    CASE WHEN (2022-(1900+year)) < 45 THEN '<45' ELSE '45+' END AS age_range
                    FROM results""", conn)
df1
```

```python
# 1.4 How many vehicles are in our dataset that are under 45 and 45 and over?
less_than_45 = pd.read_sql_query("""SELECT COUNT(year)
                FROM results
                WHERE (2022-(1900+year))<45
                   """, conn)
more_than_45 = pd.read_sql_query("""SELECT COUNT(year)
                FROM results
                WHERE (2022-(1900+year))>45
                   """, conn)
print(f"more than 45: {more_than_45}")
print(f"less than 45: {less_than_45}")
```

```python
# 1.5 List all unique vehicle name values that have 8 cylinders 
# with horsepower that is greater than 125 but doesn't exceed 200
pd.read_sql_query(""" SELECT DISTINCT(name)
                    FROM results
                    WHERE cylinders = 8 AND horsepower>125 AND horsepower<=200""",conn)

```

```python
# **2.2.3** How many song names are in this dataset that contain _'radio edit'_ in 
# their name? 
# Make sure your query is not case sensitive and there is only one single character (space)
#  between _'radio'_ and _'edit'_.


pd.read_sql_query("""SELECT count(DISTINCT song_name)
FROM df2
WHERE song_name LIKE '%radio edit%'""",conn_1)
```

```python
# **2.2.4** Write an SQL query to report songs with popularity of 99 and above. 
# How many songs are there?

pd.read_sql_query("""SELECT song_name
FROM df2
WHERE song_popularity >=99""",conn_1)
```

```python
pd.read_sql_query("""SELECT count(DISTINCT song_name)
FROM df2
WHERE song_popularity >=99""",conn_1)
```

```python
# **2.2.5** Create a new column that contains song duration in the following format m:s. 
# Example, the duration of Boulevard of Broken Dreams is 4:22


min_sec = pd.read_sql_query("""SELECT   CAST (song_duration_ms / 60000 AS STR) || ' : ' ||
                      CAST(CAST(song_duration_ms / 1000 AS INT)%60 AS STR) as time
                      
                    
                 FROM df2 """, conn_1)
min_sec
```

```python
# **2.2.6** Display the minimum, maximum and average popularity scores. 



maximum=pd.read_sql_query("""SELECT song_name, MAX(song_popularity) AS maximum_popularity
                 FROM df2""", conn_1)

minimum=pd.read_sql_query("""SELECT song_name, MIN (song_popularity) AS minimum_popularity
                 FROM df2""", conn_1)

average = pd.read_sql_query("""SELECT song_popularity, AVG (song_popularity) AS average_popularity
                 FROM df2""", conn_1)

print(f"{maximum}'\n'{minimum} '\n'{average}")
```

```python
# **2.2.7** Display average popularity scores and count for songs that contain "love" 
# in their name and those that don't. Is there any difference in their popularity? 


love = average = pd.read_sql_query(""" SELECT song_name, song_popularity 
FROM df2
WHERE song_name LIKE '%love%'""",conn_1)
not_love = pd.read_sql_query(""" SELECT song_name, song_popularity 
FROM df2
WHERE song_name NOT LIKE '%love%'""",conn_1)
not_love
average_love = pd.read_sql_query(""" SELECT AVG(song_popularity) as average_love_songs_popularity
FROM df2
WHERE song_name LIKE '%love%'""",conn_1)
average_not_love = pd.read_sql_query(""" SELECT AVG(song_popularity) as average_not_love_songs_popularity
FROM df2
WHERE song_name NOT LIKE '%love%'""",conn_1)
print(f"{average_love}'\n'{average_not_love}")
```

```python
# **2.2.8** Display top 5 songs with the longest names.
pd.read_sql_query(""" SELECT song_name , LENGTH(song_name) as len
FROM df2
ORDER BY LENGTH(song_name) DESC
LIMIT 5""",conn_1)
```
